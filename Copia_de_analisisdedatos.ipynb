```python
import miceforest as mf
from sklearn.datasets import load_iris
import numpy as np
```

```python
from sklearn import datasets
iris= datasets.load_iris()
iris
```

```python
df_iris= pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target']= iris.target
df_iris
```

```python
df_iris.rename({"target":"species"}, inplace=True, axis=1)
df_iris
```

```python
df_iris['species']= df_iris['species'].astype('category')
df_iris.info()
```

```python
######Asigna valores nulos aleatoriamente #####
iris_amp= mf.ampute_data(df_iris, perc= 0.25, random_state=1)
iris_amp
```

```python
kds= mf.ImputationKernel(
    iris_amp,
    random_state=1)

kds.mice(5)

iris_completo= kds.complete_data()
iris_completo
```

```python
# Grafico de cajas para relacion entre sexo y edad
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid', palette='Set2', font= 'sans-serif')
plt.figure(figsize=(10,10))
sns.boxplot(x=df["sexo"], y=df["edad"])
plt.xticks(rotation=90)
plt.title("Distribucion en hurto de personas para  edad por sexo", fontsize=14)
plt.xlabel("Sexo", fontsize=12)
plt.ylabel("Edad")
plt.show()
```

```python
##### Diagrama de cajas y bigotes #####
plt.figure(figsize=(10,10))
sns.boxplot(
    x=df["edad"],
    y=df["estado_civil"],
    orient="horizontal")
plt.xticks(rotation=90)
plt.title("Distribucion en hurto de personas para  edad por estado civil", fontsize=14)
plt.xlabel("estado_civil", fontsize=12)
plt.ylabel("Edad")
plt.show()

```

```python
sns.catplot(
    data=df,
    x="estado_civil",
    y="edad",
    hue="sexo",
    kind="bar")

plt.xticks(rotation=90)
plt.show()
```

```python
#### modalidad del hurto ####
plt.figure(figsize=(10,10))
sns.catplot(
    data=df,
    x="modalidad",
    y="edad",
    hue="sexo",
    kind="bar")

plt.xticks(rotation=90)
plt.show()
```
